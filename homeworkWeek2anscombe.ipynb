```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Anscombe's Quartet:
x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
plt.scatter(x, y1)
```

```python
#Calculate the slope m and intercept b of the line using np.polyfit
m, b = np.polyfit(x, y1, 1)
regression_line = m * np.array(x) + b
```

```python
plt.scatter(x,y1)
plt.plot(x, regression_line,color='red')
plt.xlabel('x')
plt.ylabel('y1')
```

```python
plt.scatter(x,y2)
plt.plot(x, regression_line,color='red')
plt.xlabel('x')
plt.ylabel('y2')
```

```python
plt.scatter(x,y3)
plt.plot(x, regression_line,color='red')
plt.xlabel('x')
plt.ylabel('y3')
```

```python
plt.scatter(x,y4)
plt.plot(x, regression_line,color='red')
plt.xlabel('x')
plt.ylabel('y4')
```

```python
plt.scatter(x4,y4)
plt.plot(x, regression_line,color='red')
plt.xlabel('x4')
plt.ylabel('y4')
```

##How do the datasets visually differ despite having similar summary statistics?

The graphs illustrate both the influence of outliers on the data and how they affect its overall distribution.

##How did the outlier in the x4, y4 dataset affect the regression line compared to the other datasets?

The slope for the regression line is steeper


##Why is it important to visualize data in addition to calculating summary statistics?

A graph makes it easier for casual viewers to visualize data compared to a table of numbers. It highlights obvious outliers and, as seen in this dataset, clearly shows how the data is distributed
