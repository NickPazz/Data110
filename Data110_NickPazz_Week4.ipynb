```python
import pandas as pd
import matplotlib.pyplot as plt

#Crime data from table

#Creating Data Dictionary
data = {'Offense': ['Aggravated Assualt', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery'],
'Per_Hundred_Thousand' : [264.1, 250.7, 1347.2, 318.7, 5.7, 38, 66.5]}

dm = pd.DataFrame(data)

#Sort the Dataframe by offense in ascending and descending order

dm_sorted = dm.sort_values(by= 'Offense', ascending=True)

#

#Plot the horizontal style bar chart
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,6))
plt.barh(dm_sorted['Offense'], dm_sorted['Per_Hundred_Thousand'], color = 'brown', zorder=2)

#Adding Labels
plt.xlabel('Crime Per 100,000')
plt.ylabel('Type of offense')
plt.title('What Crime?')

## Setting xticks
plt.xticks([0, 400, 800, 1200, 1500])

#Customizing ticks (Style)

ax = plt.gca()
ax.tick_params(axis='y', labelsize=8)
plt.grid(axis='y', linestyle='-', color= 'gray', zorder = 0)
plt.tight_layout()



#Defining custom list of colors
custom_colors = ['#FFB6C1', '#FA8072', '#FF0000', '#8B0000', '#DC143C', '#B22222', '#CD5C5C']
plt.barh(dm_sorted['Offense'], dm_sorted['Per_Hundred_Thousand'], color=custom_colors, zorder=2)
```

```python
##Displaying crime change with Pandas. Pandas is superior.

data = {'Offense': ['Aggravated Assualt', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery'],
         'Change' : [-3.3, -8.1, -4.9, 12, -12, -9.8, -.8]}

df = pd.DataFrame(data)

df_sorted = df.sort_values(by= 'Offense', ascending=True)

#Ccustom color map

custom_colors = ['#FFB6C1', '#FA8072', '#FF0000', '#8B0000', '#DC143C', '#B22222', '#CD5C5C']

plt.figure(figsize=(10,6))
plt.barh(df['Offense'], df['Change'], color =custom_colors)

plt.xlabel('Percentage Change')
plt.ylabel('Type of offense')
plt.title('Percentage Change in Crime Offenses 2023-2024')
plt.tight_layout()

#customizing ticks
plt.xticks([-15, -10, 0, 10, 15])

# Remove specific spines (borders)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

ax= plt.gca()
ax.tick_params(axis='y', labelsize=10)
plt.grid(axis='y', linestyle='-', color='gray', zorder=0)
plt.tight_layout()


```

**Explain why seeing the change over time is more insightful than just seeing the total number of offenses for a single year**

The data could be interpreted show significant changes in crime trends post-pandemic. However, the shifts between the displayed years may appear more drastic than what is actually occurring.

The stacked bar chart below illustrates crime data for 2022 and 2023. The graph above displays the percentage change in crime, which may appear larger than the actual total change in crime represented by the stacked bar chart.

```python
import numpy as np

##stacked bar data

x = ['Aggravated Assualt', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery']
y = [273.2, 272.1, 1416.4, 284.5, 6.48, 42.1, 67.01]
z = [264.1, 250.7, 1347.2, 318.7, 5.7, 38, 66.5]

#Generate index from x-axis
x_indexes = np.arange(len(x))

plt.bar(x_indexes - 0.125, y, width=0.25, label='2022')
plt.bar(x_indexes + 0.125, z, width=0.25, label='2023')

plt.xticks(ticks=x_indexes, labels=x)


#Add titles and lables
plt.title('Crime rate change between 2022-2023')
plt.xlabel('Crime')
plt.ylabel('Crime per 100,000')

plt.xticks(ticks=x_indexes, labels=x)

plt.legend()
plt.tight_layout()
```
